```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
sns.set_style('whitegrid')
```

```python
credit_df = pd.read_csv("../data/raw/creditcard.csv")
```

```python
credit_df.head()
```

```python
credit_df.shape
```

```python
credit_df.info()
```

```python
credit_df.describe()
```

```python
credit_df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(credit_df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

Target Variable Analysis

```python
credit_df['Class'].value_counts()
```

```python
credit_df['Class'].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Class',
    data=credit_df
)

plt.title('Class Distribution')
plt.show()
```

```python
fraud_rate = (
    credit_df['Class'].mean()
    * 100
)

print(f"Fraud Rate: {fraud_rate:.4f}%")
```

Univariate Analysis

```python
plt.figure(figsize=(8,5))

sns.histplot(
    credit_df['Amount'],
    bins=50,
    kde=True
)

plt.title('Transaction Amount Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(
    x=credit_df['Amount']
)

plt.title('Transaction Amount Boxplot')
plt.show()
```

Bivariate Analysis

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Class',
    y='Amount',
    data=credit_df
)

plt.title('Amount by Transaction Class')
plt.show()
```

```python
fraud_corr = (
    credit_df.corr()['Class']
    .sort_values(ascending=False)
)
```

```python
fraud_corr
```

```python
fraud_corr.head(10)
```

```python
fraud_corr.tail(10)
```

```python
plt.figure(figsize=(8,10))

fraud_corr.drop('Class').sort_values().plot(
    kind='barh'
)

plt.title('Feature Correlation with Fraud')
plt.show()
```
