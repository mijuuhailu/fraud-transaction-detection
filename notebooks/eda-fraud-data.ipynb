Target Variable Analysis

```python
fraud_df['class'].value_counts()
```

```python
fraud_df['class'].value_counts(normalize=True)*100
```

```python
sns.countplot(
    x='class',
    data=fraud_df
)

plt.title("Fraud Class Distribution")
plt.show()
```

Univariate Analysis

```python
numerical_cols = [
    'purchase_value',
    'age'
]
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(fraud_df[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
categorical_cols = [
    'source',
    'browser',
    'sex'
]
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(
        x=col,
        data=fraud_df
    )
    plt.xticks(rotation=45)
    plt.show()
```

Time-Based Exploration

```python
fraud_df['signup_time'] = pd.to_datetime(
    fraud_df['signup_time']
)

fraud_df['purchase_time'] = pd.to_datetime(
    fraud_df['purchase_time']
)
```

```python
fraud_df['hour_of_day'] = (
    fraud_df['purchase_time']
    .dt.hour
)
```

```python
sns.countplot(
    x='hour_of_day',
    data=fraud_df
)

plt.xticks(rotation=90)
plt.show()
```

Bivariate Analysis

```python
sns.boxplot(
    x='class',
    y='purchase_value',
    data=fraud_df
)
```

```python
pd.crosstab(
    fraud_df['browser'],
    fraud_df['class']
)
```

```python
browser_fraud = pd.crosstab(
    fraud_df['browser'],
    fraud_df['class']
)

sns.heatmap(
    browser_fraud,
    annot=True,
    fmt='d'
)
```
